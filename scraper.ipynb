```python
import datetime
import pandas as pd
import time
import requests
import os

API_KEY = "_T8YnZaxCVsyj_AKsnus"

URL_TEMPLATE = "https://www.quandl.com/api/v3/datasets/EOD/{0}.csv?api_key={1}"
URL_TEMPLATE = "https://www.quandl.com/api/v3/datatables/WIKI/PRICES?date.gte={}&date.lt={}&ticker={}&api_key={}"

START_DATE = "2000-01-01"
END_DATE = "2017-11-01"
```

```python
dj_components = [
    "DD",
    "UNH",
    "CSCO",
    "HD",
    "JNJ",
    "PG",
    "GE",
    "TRV",
    "AXP",
    "MRK",
    "NKE",
    "MSFT",
    "UTX",
    "PFE",
    "GS",
    "V",
    "IBM",
    "JPM",
    "WMT",
    "XOM",
    "KO",
    "MMM",
    "MCD",
    "CAT",
    "INTC",
    "DIS",
    "BA",
    "AAPL",
    "CVX",
    "VZ"
]

directory = "djia_20000101_20171101"
```

```python

if not os.path.exists(directory):
    os.makedirs(directory)

for ticker in dj_components:
    url = URL_TEMPLATE.format(START_DATE, END_DATE, ticker, API_KEY)
    r = requests.get(url).json()
    col_names = [c["name"] for c in r["datatable"]["columns"]]
    df = pd.DataFrame(r["datatable"]["data"], columns = col_names)
    
    df = df[((df.date >= START_DATE) & (df.date <= END_DATE))]
    df = df[["date", "adj_open", "adj_high", "adj_low", "adj_close", "adj_volume"]]
    df.to_csv("{}/{}.csv".format(directory, ticker.lower()), index=False)
```
